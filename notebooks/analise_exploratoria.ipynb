```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('D:\wealthpath-advisor\data\data\simulacao_clientes.csv')

#Verificar tipos de dados e nulos
print(df.info())

#Estatística Descritiva
print(df.describe())
```

```
#Qual a média de Ativos dos Clientes que estamos prospectando?
plt.figure(figsize=(10,6))
sns.histplot(df['Patrimonio_Inicial'], kde=True, color='gold')
plt.title('Distribuição de Patrimonio Bruto - Base HNW')
plt.show()

#Existe relação entre o Perfil de Risco e o Aporte Mensal?
sns.boxplot(x='Perfil_Risco', y='Gasto_Mensal_Desejado', data=df)
plt.title('Estilo de Vida vs. Perfil de Risco')
plt.show()
```

```
# Configurando o estilo
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 5))

# Gráfico 1: Histograma de Idade vs Aposentadoria
plt.subplot(1, 2, 1)
sns.histplot(df['Idade'], kde=True, color='blue', label='Idade Atual')
sns.histplot(df['Idade_Aposentadoria'], kde=True, color='green', label='Aposentadoria')
plt.title('Janela de Acumulação Patrimonial')
plt.legend()

# Gráfico 2: Relação Aporte vs Gasto por Perfil de Risco
plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x='Aporte_Mensal', y='Gasto_Mensal_Desejado', hue='Perfil_Risco', palette='viridis', s=100)
plt.title('Capacidade de Poupança por Perfil')

plt.tight_layout()
plt.show()
```
